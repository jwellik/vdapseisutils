### Download data for a Mt. Hood/Wy'East earthquake

```python
from seismology.stream.archive import archive
from ioutils.filestructure import sds_starttime

#ds             = 'wws:130.118.152.200:16022'  # DOI VPN required
ds             = 'IRIS'

nslc_list = [
    'CC.PALM..*',
    'CC.TIMB..*',
    'CC.YOCR..*',
    'CC.BRSP..*',
    'CC.LSON..*',
    'CC.HIYU..*',
    'CC.SHRK..*',
    'UW.HOOD..*',
    'UW.TDH..EHZ',
    'UW.VLL..EHZ',
]

tstart        = '2021/10/18 00:00:00'
tend          = '2021/10/19 00:00:00'

st = archive(ds, nslc_list, tstart, tend, basedir='/Users/jjw2/Desktop/MtHood/', filestructure=sds_starttime, reclen=512, return_stream=True)
```
